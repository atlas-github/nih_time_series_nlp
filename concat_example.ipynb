```python
#download dengue csv
import pandas as pd
import gdown

# insert file_id
file_id1 = '1F-faNnQoyhdjbuyEVZPHV1cv_h0h5UDl'
url1 = f'https://drive.google.com/uc?id={file_id1}'

# Download the CSV file
gdown.download(url1, 'dengue.csv', quiet=False)
```

```python
# Read the CSV file into a DataFrame
df_dengue1 = pd.read_csv('dengue.csv')
df_dengue1
```

```python
# insert file_id
file_id2 = '101YUU3obMz_qlsrPLk3TrCreMvcLaMEx'
url2 = f'https://drive.google.com/uc?id={file_id2}'

# Download the XLSX file
gdown.download(url2, 'data 2014.xlsx', quiet=False)
```

```python
# Read the XLSX file into a DataFrame
df_dengue2 = pd.read_excel('data 2014.xlsx', sheet_name = "MAIN DATA", dtype={'Rapid Test - Tkh Ujian Keputusan': str})
df_dengue2
```

```python
# concatenate (stack dataframes vertically, and fill missing values with NaN)
df_dengue_concat = pd.concat([df_dengue1, df_dengue2], ignore_index = True)
df_dengue_concat
```
