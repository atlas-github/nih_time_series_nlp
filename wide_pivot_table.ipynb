```python
#download dengue csv
import pandas as pd
import gdown

# insert file_id
file_id = '1F-faNnQoyhdjbuyEVZPHV1cv_h0h5UDl'
url = f'https://drive.google.com/uc?id={file_id}'

# Download the CSV file
gdown.download(url, 'dengue.csv', quiet=False)

# Read the CSV file into a DataFrame
df_dengue = pd.read_csv('dengue.csv')
df_dengue
```

```python
df_dengue.head(10).to_csv("df_dengue.csv")

from google.colab import files
files.download("df_dengue.csv")
```

```python
pivot_df = pd.pivot_table(
    df_dengue,
    values='NO_KES',                # The value to aggregate (could also use 'count')
    index='TRK_NOTI',                   # Rows - Date
    columns=['JNS_KES', 'DAERAH'],  # Columns for 'type_of_case' and subcolumns for 'district'
    aggfunc='count',                # Aggregation function
    fill_value=0                    # Fill missing values with 0
)

pivot_df
```

```python
pivot_df.to_csv("pivot_df.csv")

from google.colab import files
files.download("pivot_df.csv")
```
